# Descrição do Problema

O turismo é uma atividade econômica e social de grande relevância em escala global, movimentando milhões de pessoas e gerando impacto significativo na economia de diversos países. Com o aumento da mobilidade internacional, torna-se essencial compreender os fatores que influenciam a popularidade de destinos turísticos ao redor do mundo.

Neste contexto, este projeto tem como objetivo realizar uma Análise Exploratória de Dados (EDA) a partir de um conjunto de dados contendo informações sobre locais turísticos famosos, incluindo características como país, cidade, número de visitantes, tipo de atração, taxa de entrada e receita gerada.

A análise busca identificar padrões, tendências e possíveis relações entre essas variáveis, respondendo questões como: quais fatores estão associados à popularidade dos destinos, como os tipos de atrações se distribuem globalmente e de que forma variáveis econômicas, como receita e custo, se relacionam com o fluxo de visitantes.

# Hipóteses do Problema

As hipóteses consideradas para esta análise são as seguintes:

1. Destinos turísticos com maior número de visitantes tendem a gerar maior receita.

## Normalização

A normalização escala os dados para um intervalo fixo, geralmente entre 0 e 1. É útil quando o algoritmo de machine learning assume que as características estão em uma escala semelhante.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
url = "https://raw.githubusercontent.com/milcagomescoutinho/exploratory-data-analysis-tourism/main/world_famous_places_2024.csv"
df = pd.read_csv(url)
```

```python

from sklearn.preprocessing import MinMaxScaler
```

```python
X = df[['Annual_Visitors_Millions',
        'Entry_Fee_USD',
        'Average_Visit_Duration_Hours']]

y = df['Tourism_Revenue_Million_USD']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
scaler_norm = MinMaxScaler()

X_train_normalized = scaler_norm.fit_transform(X_train)
X_test_normalized = scaler_norm.transform(X_test)
```

```python
import pandas as pd

df_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)
df_normalized.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))

sns.histplot(df_normalized['Annual_Visitors_Millions'], kde=True)

plt.title('Distribuição do Número de Visitantes (Normalizado)')
plt.xlabel('Visitantes (normalizado)')
plt.ylabel('Frequência')

plt.show()
```

O histograma dos dados normalizados evidencia que a distribuição do número de visitantes permanece assimétrica, com maior concentração de valores em níveis mais baixos e poucos destinos apresentando valores elevados.

Isso indica que a maioria dos locais possui um volume moderado de visitantes, enquanto poucos destinos concentram grande fluxo turístico.

A normalização preserva o padrão da distribuição original, permitindo melhor comparação entre variáveis em diferentes escalas.

## Padronização

A padronização (ou Z-score scaling) transforma os dados para ter média 0 e desvio padrão 1. É útil para algoritmos que são sensíveis à escala das características, como SVMs ou redes neurais.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_std = StandardScaler()

X_train_standardized = scaler_std.fit_transform(X_train)
X_test_standardized = scaler_std.transform(X_test)
```

```python
df_standardized = pd.DataFrame(X_train_standardized, columns=X_train.columns)

df_standardized.head()
```

```python
plt.figure(figsize=(8, 6))

sns.histplot(df_standardized['Annual_Visitors_Millions'], kde=True)

plt.title('Distribuição do Número de Visitantes (Padronizado)')
plt.xlabel('Visitantes (padronizado)')
plt.ylabel('Frequência')

plt.show()
```

O histograma dos dados padronizados mostra que a variável de número de visitantes foi centralizada em torno de zero, com valores distribuídos tanto no sentido negativo quanto positivo.

Observa-se que a forma da distribuição foi preservada, mantendo a assimetria original dos dados, porém agora em uma escala padronizada.

Essa transformação é importante para garantir que as variáveis tenham média zero e desvio padrão igual a um, evitando que diferenças de escala impactem análises e modelos futuros.

## Outras Transformações e Etapas de Pré-Processamento

Além das etapas realizadas, poderiam ser exploradas técnicas mais avançadas de pré-processamento, como seleção de variáveis baseada em correlação ou importância estatística, reduzindo a complexidade do modelo.

A aplicação de PCA (Análise de Componentes Principais) também poderia ser utilizada para reduzir a dimensionalidade dos dados, preservando a maior parte da variância explicativa.

Adicionalmente, técnicas de feature engineering poderiam gerar novos atributos mais representativos, como métricas de eficiência turística (receita por visitante) ou segmentações por região.

Essas abordagens contribuiriam para uma modelagem mais robusta, interpretável e com maior capacidade preditiva.


```python
from sklearn.decomposition import PCA

X_pca = df[['Annual_Visitors_Millions',
            'Entry_Fee_USD',
            'Tourism_Revenue_Million_USD',
            'Average_Visit_Duration_Hours']]

pca = PCA(n_components=2)
X_pca_result = pca.fit_transform(X_pca)

df_pca = pd.DataFrame(X_pca_result, columns=['PC1', 'PC2'])

df_pca.head()
```

```python
print("Variância explicada:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df_pca['PC1'], df_pca['PC2'])

plt.title('Visualização dos Dados com PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')

plt.show()
```

A visualização dos dados utilizando PCA permite observar a distribuição dos destinos turísticos em um espaço bidimensional, baseado nas componentes principais.

Nota-se que a maioria dos pontos está concentrada em uma região específica, indicando similaridade entre diversos destinos. No entanto, alguns pontos aparecem mais distantes, sugerindo a presença de destinos com características significativamente diferentes, possivelmente com maior volume de visitantes ou receita.

Essa representação evidencia a capacidade do PCA de reduzir a dimensionalidade dos dados, mantendo as principais variações e facilitando a identificação de padrões e outliers.
